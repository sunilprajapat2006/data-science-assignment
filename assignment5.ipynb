Pandas is an open-source Python library built on top of NumPy that provides fast, flexible, and expressive
data structures designed to work with structured (tabular) data. The name comes from Panel Data, a term
used in econometrics. It was created by Wes McKinney in 2008 and is now one of the most popular tools
in the data science ecosystem.

Role in Data Analysis:
•  Data Loading: Easily read data from CSV, Excel, SQL, JSON, HTML, and more using functions like
pd.read_csv(), pd.read_excel().
•  Data Cleaning: Handle missing values (fillna(), dropna()), remove duplicates, correct data types.
•  Data Exploration: Use head(), tail(), info(), describe() to quickly understand dataset structure and
statistics.
•  Data Manipulation: Filter rows, select columns, sort, group, merge, and reshape data.
•  Data Aggregation: Use groupby() for group-level statistics like sum, mean, count.
•  Time-Series Analysis: Powerful date/time indexing for time-series data.

Role in Machine Learning:
•  Data Preprocessing: Prepare raw data for ML models — normalize, encode categorical variables,
split features and labels.
•  Feature Engineering: Create new features from existing columns.
•  Integration: Seamlessly works with Scikit-learn, TensorFlow, and PyTorch. DataFrames can be
converted to NumPy arrays required by ML libraries.
•  Pipeline Building: Acts as the backbone of data pipelines in MLops workflows.

```python
import pandas as pd
# Create the DataFrame
data = {
    'Name':  ['Amit', 'Neha', 'Rahul', 'Sneha'],
    'Age':   [20, 21, 19, 22],
    'Marks': [85, 90, 78, 88]
}
df = pd.DataFrame(data)
# Task 1: Display first 2 rows
print("First 2 rows:")
print(df.head(2))
# Task 2: Display only Name and Marks columns
print("\nName and Marks columns:")
print(df[['Name', 'Marks']])
# Task 3: Add a new column 'Result'
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x >= 40 else 'Fail')
print("\nDataFrame with Result column:")
print(df)
# Task 4: Display students with Marks > 80
print("\nStudents with Marks > 80:")
print(df[df['Marks'] > 80])
```

```python
import pandas as pd
import numpy as np
# Create the DataFrame with missing values
data = {
    'Name':   ['A', 'B', 'C', 'D'],
    'Age':    [25, np.nan, 27, 24],
    'Salary': [30000, 40000, np.nan, 35000]
}
df = pd.DataFrame(data)
# Task 1: Identify missing values
print("Missing values in each column:")
print(df.isnull().sum())
print("\nMissing value positions (True = missing):")
print(df.isnull())
# Task 2: Fill missing Age with average age
avg_age = df['Age'].mean()
print(f"\nAverage Age: {avg_age}")
df['Age'] = df['Age'].fillna(avg_age)
# Task 3: Fill missing Salary with median salary
med_salary = df['Salary'].median()
print(f"Median Salary: {med_salary}")
df['Salary'] = df['Salary'].fillna(med_salary)
# Task 4: Display cleaned dataset
print("\nCleaned Dataset:")
print(df)
```

```python
import pandas as pd
# Load the Iris dataset from URL
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)
# Task 1: Display dataset information
print("Dataset Info:")
df.info()
# Task 2a: Total rows and columns
print(f"\nTotal Rows: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
print(f"Shape (rows x cols): {df.shape}")
# Task 2b: Average sepal length
avg_sepal = df['sepal_length'].mean()
print(f"\nAverage Sepal Length: {avg_sepal:.2f} cm")
# Task 2c: Different species names
print(f"\nDifferent Species Names:")
print(df['species'].unique())
print(f"Total unique species: {df['species'].nunique()}")
# Task 3: Display summary statistics
print("\nSummary Statistics:")
print(df.describe())
```
